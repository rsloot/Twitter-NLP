```python
import twitter
import json
import numpy as np
import matplotlib.pyplot as plt

import time
import datetime

import string
import nltk
from nltk import PorterStemmer
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from auth_twitter import auth_twit
```

```python
twittur = auth_twit(app_name='treeing').auth()
```

```python
count = 0
data = []
while count < 5:
    data.append(twittur.statuses.home_timeline(count=10))
    count += 1
    time.sleep(300)
```

```python
for i in range(len(data)):
    for j in range(len(data[i])):
        print('-',data[i][j]['text'])
```

```python
all_text = ' '.join([data[i][j]['text'] for i in range(len(data)) for j in range(len(data[i]))]).split()
```

```python
from collections import Counter
Counter(' '.join(all_text).split()).most_common
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
stop_words
```

```python
for i,word in enumerate(all_text):
    if word in stop_words:
        all_text.pop(i)
```
