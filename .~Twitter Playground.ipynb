```python
stop_words = nltk.corpus.stopwords.words('english') + nltk.corpus.stopwords.words('spanish') + list(string.punctuation) + ['rt']
```

```python
# print(stop_words)
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
loops = 0
count = -1

for trend, counter1 in word_counts.items():
    for k_word, word_count in counter1.most_common():

        if len(k_word) < 2:
            word_counts[trend].pop(k_word)
            count += 1
            continue
        if ps.stem(k_word.lower()) in stop_words:
            word_counts[trend].pop(k_word)
            count += 1
        elif k_word.lower() in stop_words:
            word_counts[trend].pop(k_word)
            count += 1

print('{0} words removed'.format(count))
# commons_25th = [i for i in c if i not in stop_words]
```

```python
## combine case variation for words
# word_counts
## {trend: {Counter: {'words': word_count}}...}
for key_trend, v_counter in word_counts.items():
    
    for key_word, word_count in v_counter.most_common():

        if key_word[-1] in string.punctuation:
            if key_word[:-1] in v_counter:
                word_counts[key_trend][key_word[:-1]] += word_counts[key_trend].pop(key_word)

            else:
                word_counts[key_trend][key_word[:-1]] = word_counts[key_trend].pop(key_word)

# temp_word_counts = word_counts.copy()
for key_trend, v_counter in word_counts.items():
    
    for key_word, word_count in v_counter.most_common():
        if (key_word != key_word.lower() and key_word.lower() in v_counter):
            word_counts[key_trend][key_word] += word_counts[key_trend].pop(key_word.lower())
        
        

#         else:
#             mentions[k][word] = word_counts[k].pop(word)        
```

```python
def prettyprint_counts(trend, label, list_of_tuples):
#     print(trend)
    print("\n{:^25} | {:^6}".format(label, "Count"))
    print("*"*40)
    for k,v in list_of_tuples:
        print("{:25} | {:>6}".format(k,v))
```
