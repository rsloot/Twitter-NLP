```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
%matplotlib inline
```

```python
with open('./Monthly_trends/October.txt', 'r+') as f:
    d = json.loads(f.read())
```

```python
trends_dict = {}
for i in d:
    for k,v in d[i].items():
        if k in trends_dict:
            trends_dict[k] += v
        else:
            trends_dict[k] = v
#     break
```

```python
from collections import OrderedDict
sorted_trends = OrderedDict(sorted(trends_dict.items(),
                                   key=lambda x: x[1],
                                   reverse=True)[:100])#top 10
sorted_trends
```
